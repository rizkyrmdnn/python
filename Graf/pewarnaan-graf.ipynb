```python
# Langkah 2: Import library yang dibutuhkan
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Langkah 3: Membuat graf sederhana
G = nx.Graph()
```

```python
# Menambahkan simpul dan sisi (edges)
edges = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4)]
G.add_edges_from(edges)
```

```python
# Langkah 4: Algoritma pewarnaan graf (greedy coloring)
coloring = nx.coloring.greedy_color(G, strategy="largest_first")
```

```python
# Langkah 5: Visualisasi graf dengan pewarnaan
# Menentukan warna untuk setiap node berdasarkan hasil pewarnaan
colors = [coloring[node] for node in G.nodes()]
```

```python
# Menggambar graf dengan pewarnaan
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color=colors, node_size=800, cmap=plt.cm.Set3, font_size=16, font_color='black')
plt.title("Pewarnaan Graf dengan Greedy Algorithm", size=15)
plt.show()
```

```python
# Langkah 6: Menampilkan hasil pewarnaan
print("Pewarnaan tiap simpul:")
for node, color in coloring.items():
    print(f"Simpul {node} diberi warna {color}")
```
